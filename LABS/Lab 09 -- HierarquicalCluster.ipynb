```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist
```

```python
data = np.random.rand(6, 2)
```

```python
print(data)
```

```python
# Calcular as distâncias entre os pontos
dist_matrix = pdist(data, metric='euclidean')
# Executar o algoritmo de aglomeração hierárquica usando a ligação simples(min)
Z = linkage(dist_matrix, method='single')
```

```python
# Plotar o dendrograma
plt.figure(figsize=(9, 6))
dendrogram(Z)
plt.title('Dendrograma - Aglomeração Hierárquica (Ligação Simples)')
plt.xlabel('Índice da Amostra')
plt.ylabel('Distância')
plt.show()
```
