## Exercise 3 - Implement Fisher's linear discriminant

* Use numpy to calculate $\vec{w}$.
* Plot the data (as you did in Exercise 1), and add a line on the direction of $\vec{w}$ (reference Figure below).
* Project the points onto the calculated vector, and plot the projection (as you did in Exercise 2) (reference Figure below).

```python
# your code goes here:
def FisherLinearDiscriminant(X,y):
    '''
    X: np.array 2dim, com 2 colunas feature 1 e feature 2
    y: np.array 1 dim, com as classes respectivas da linha i de X
    '''
    x1 = X[y == 0]
    x2 = X[y == 1]

    Mean1 = np.mean(x1, axis=0)
    Mean2 = np.mean(x2, axis=0)
    Mean_difference = Mean1-Mean2

    cov1 = np.zeros((x1.shape[1], x1.shape[1]))
    for i in range(x1.shape[0]):
        cov1 += (x1[i, :] - Mean1).reshape(-1,1) @ (x1[i, :] - Mean1).reshape(-1,1).T
    cov1 /= x1.shape[0] - 1

    cov2 = np.zeros((x2.shape[1], x2.shape[1]))
    for i in range(x2.shape[0]):
        cov2 += (x2[i, :] - Mean2).reshape(-1,1) @ (x2[i, :] - Mean2).reshape(-1,1).T
    cov2 /= x2.shape[0] - 1 
    
    cov1_stable = cov1 + 1e-8 * np.eye(cov1.shape[0])
    cov2_stable = cov2 + 1e-8 * np.eye(cov2.shape[0])
    
    Sw_inv = np.linalg.inv(cov1+cov2)

    w = np.dot(np.eye(2),Mean_difference)
    return w
```

```python
def plot_reta(vetor_diretor, xrange=(-2, 5)):
    a, b = vetor_diretor
    x = np.linspace(xrange[0], xrange[1], 10)
    y = a * x + b
    plt.plot(x, y,linestyle='--',color="red")
```

```python
v_w = FisherLinearDiscriminant(X,y)

fig, ax = plt.subplots()
scatter = ax.scatter(X[:, 0], X[:, 1], c=y, label='Pontos')
legend1 = ax.legend(*scatter.legend_elements(), loc="lower left", title="Classes")
ax.add_artist(legend1)

plot_reta(v_w)
plt.grid(True)
plt.show()
```
